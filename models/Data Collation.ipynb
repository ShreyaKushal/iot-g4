```python
import pandas as pd
```

```python
existing_data = "../data/sensor1.csv"
new_data = "../data/-data-2024-03-15 18_36_10.csv" #TO-CHANGE
```

```python
df1 = pd.read_csv(existing_data)
df1 = df1.dropna()
df1["Time"] = pd.to_datetime(df1["Time"])
```

```python
df2 = pd.read_csv(new_data)
df2 = df2.dropna()
df2["Time"] = pd.to_datetime(df2["Time"])
```

```python
result = pd.concat([df1, df2], ignore_index=True).drop_duplicates(subset='Time', keep='first')
```

```python
result = result.sort_values(by='Time', ascending=True).reset_index(drop=True)
```

```python
result.to_csv('../data/sensor1.csv', index=False)
```
